```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset('iris')
```

```python
print(iris.head())
```

```python
print(iris.shape)
```

```python
sns.scatterplot(x='sepal_length', y='petal_length', hue='species', data=iris)

plt.show()
```

Conclusion:
Different flower species show different sepal and petal measurements.
Boxplots help identify spread and outliers in data.
